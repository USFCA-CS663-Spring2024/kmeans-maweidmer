## KMeans Implementation Performance Comparison

```python
from KMeans import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics.cluster import v_measure_score
from sklearn.cluster import KMeans as SKmeans
import seaborn as sns
import pandas as pd
```

#### Generate the data points to use for the following comparisons

```python
X, cluster_assignments = make_blobs(n_samples=700, centers=4, cluster_std=0.60, random_state=0)
```

```python
X
```

#### Creating a pandas dataframe to hold the values data points, for easier plotting later on.

```python
df = pd.DataFrame({'X': X[:, 0], 'Y': X[:, 1]})
```

```python
df
```

#### Initialise my implementation of KMeans with 4 target clusters to match make_blob, and then use my implementation of fit to get my hypotheses.

```python
kmeans = KMeans(4)
k_labels, k_centroids = kmeans.fit(X)
```

#### The centroids found by my implementation

```python
k_centroids
```

#### The hypotheses assigned by my implementation

```python
k_labels
```

#### Generating cluster hypotheses using sklearn's implementation of KMeans

```python
skmeans = SKmeans(n_clusters=4)
skmeans.fit(X)
```

#### The centroids generated by scikit-learns implementation

```python
skmeans.cluster_centers_
```

#### The hypotheses assigned by scikit-learns implementation

```python
skmeans.labels_
```

#### Adding all of the different labels to the pandas dataframe for plotting purposes

```python
df['clusters'] = cluster_assignments
df['kmeans_hypotheses'] = k_labels
df['skmeans_hypotheses'] = skmeans.labels_
df
```

#### The data from make_blobs, shown with the clusters as generated by make_blobs

```python
sns.scatterplot(df, x=df.X, y=df.Y, hue=df.clusters).set_title('Instances generated from make_blobs')
```

#### The data from make_blobs, coloured according to the hypotheses generated by my implementation of KMeans.

The points are generally clustered the same, although one point within cluster 0 (as labeled by make_blobs) is said to be in cluster 2 (as labeled by make_blobs).

```python
sns.scatterplot(df, x=df.X, y=df.Y, hue=df.kmeans_hypotheses).set_title('Instances with clusters generated by my KMeans')
```

As a metric to measure the performance of my implementation I decided to use V-Measure, as was mentioned in the recorded lecture, which combines measurements of cluster "purity" (homogeneity), and completeness.  I decided to use V-Measure as we have the labels from make_blobs, and as mentioned in the lecture sklearn includes an implementation which I was able to use.

```python
v_measure_score(cluster_assignments, k_labels)
```

As can be seen, my v-measure score is 0.99, where 1 indicates perfect homogeneity and completeness, indicating good performance.

#### Comparing to scikit-learns implementation of KMeans

As can be seen in the graph, scikit-learns implementation also labels the same point I mentioned previously as being in cluster 2 instead of cluster 0.  This leads to scikit-learn's KMeans receiving the same v-measure score of 0.99 when measured against the cluster_assignments from make_blob.

```python
v_measure_score(cluster_assignments, skmeans.labels_)
```

```python
sns.scatterplot(df, x=df.X, y=df.Y, hue=df.skmeans_hypotheses).set_title('Instances with clusters generated by scikit-learns KMeans')
```

When comparing my implementation's hypotheses against scikit-learn's, the v-measure score is actually 1.0, indicating the cluster hypotheses for the data points are the same for scikit-learn and my implementation.

```python
v_measure_score(k_labels, skmeans.labels_)
```

### Chicago Taxi Data

```python
chicago = pd.read_csv('Taxi_Trips__2013-2023_.csv')
chicago
```

#### As suggested in the assignment spec, for the Chicago data I decided to focus on the pickup coordinate data.

```python
chicago = chicago[['Pickup Centroid Longitude', 'Pickup Centroid Latitude']]
chicago
```

Looking at the dataframe and results for .info() shows that there are missing values for coordinates.

```python
chicago.info()
```

I decided to drop the rows with missing data as I'm only looking at the two columns (longitude, latitude), so there's not really a point to consider them if both values for both columns are missing.

```python
chicago = chicago.dropna()
chicago
```

Plotting longitude against latitude reveals an impression of a map of Chicago, with more pickups centered downtown, and less in the surrounding suburbs.

https://www.google.com/search?q=chicaga+map&sourceid=chrome&ie=UTF-8

```python
sns.scatterplot(chicago, x=chicago['Pickup Centroid Longitude'], y=chicago['Pickup Centroid Latitude']).set_title('Longitude vs Latitude for Chicago Pickup Data')
```

Although I can see a cluster in the top right corner, along the waterline of Chicago, it's hard to visually pick out a good value for the number of clusters, so I decided to try using the elbow method like we did in class.

```python
inertias = pd.DataFrame(data=[], index=range(2,20), columns=['inertia'])

for k in range(2, 20):
    kmeans = SKmeans(n_clusters=k)
    kmeans.fit(chicago)
    kmeans.inertia_
    inertias.loc[k] = kmeans.inertia_

inertias

inertias.plot()
```

There is a definite elbow around 5 clusters, so I decided to use that as my hyperparameter input to KMeans.

```python
kmeans = SKmeans(5)
kmeans.fit(chicago)
chicago['kmeans_clusters'] = kmeans.labels_
sns.scatterplot(chicago, x=chicago['Pickup Centroid Longitude'], y=chicago['Pickup Centroid Latitude'], hue=chicago.kmeans_clusters).set_title('Chicago Pickup Location Data Clustered using KMeans')
```

As can be seen on the plot, KMeans indicates 3 different clusters in the suburbs west, south, and southwest, as well as two different clusters within the more densely packed downtown part of Chicago.

If I'd been able to get DBSCAN working I would have expected a more shaped cluster in the downtown area, although the surrounding clusters in the suburbs likely would have just been marked as outliers given the issues mentioned in class about DBSCAN in relation to differences in density.

Source: https://cs.uef.fi/mopsi/data/

```python
sns.scatterplot(mopsi, x=mopsi[1], y=mopsi[0]).set_title('Mopsi data')
```

Plotting the two columns against each other in the way I did makes them match up to the image on the Mopsi dataset site which shows a map of Finland, with red marked positions.

Again, I'm not sure just from looking how many clusters I should use, so I'll use the elbow method as before.

```python
inertias = pd.DataFrame(data=[], index=range(2,20), columns=['inertia'])

for k in range(2, 20):
    kmeans = SKmeans(n_clusters=k)
    kmeans.fit(mopsi)
    kmeans.inertia_
    inertias.loc[k] = kmeans.inertia_

inertias

inertias.plot()
```

Again, the elbow looks to be around 5, so I'll use that as my input to KMeans.

```python
kmeans = SKmeans(5)
kmeans.fit(mopsi)
mopsi['kmeans_clusters'] = kmeans.labels_
sns.scatterplot(mopsi, x=mopsi[1], y=mopsi[0], hue=mopsi.kmeans_clusters).set_title('Finland Location Data Clustered using KMeans')
```

Running KMeans revealed some clusters that made sense to me, specifically the the upper cluster and the cluster on the left.  However I did notice the lines of point sprinkled throughout the data (presumably roads or something like that), which I though might be better captured with DBSCAN, which allows for non-spheroidal clusters.

```python
from sklearn.cluster import DBSCAN

db_cluster = DBSCAN(eps=2, min_samples=1).fit(mopsi[[0,1]])
mopsi['dbscan_clusters'] = db_cluster.labels_
sns.scatterplot(mopsi, x=mopsi[1], y=mopsi[0], hue=mopsi.dbscan_clusters).set_title('Finland Location Data Clustered using DBSCAN')
```

Using DBSCAN did capture some of the long thin clusters as I was expecting.

One point, I ended up settling on low min_sample and eps values for the hyperparameters, as larger values ended up including all the points within a few large clusters, except for strange mini clusters which seem like they shouldn't exist, which was basically a useless result.

```python
db_cluster = DBSCAN(eps=5, min_samples=2).fit(mopsi[[0,1]])
mopsi['dbscan_clusters'] = db_cluster.labels_
sns.scatterplot(mopsi, x=mopsi[1], y=mopsi[0], hue=mopsi.dbscan_clusters).set_title('Finland Location Data Clustered using DBSCAN with larger eps and min_sample values')
```
